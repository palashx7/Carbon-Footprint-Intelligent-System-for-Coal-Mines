```python
import pandas as pd

df = pd.read_csv("carbon_neutrality_200k.csv")

# Convert the first 3 columns to numeric (force errors to NaN)
for col in df.columns[:3]:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Check if conversion worked
print(df.dtypes.head(3))
```

```python
# Check value ranges
print(df.describe())

# Check unrealistic values
invalid = df[(df.iloc[:, 2] > df.iloc[:, 0]) | (df.iloc[:, 2] < 0)]
print(f"Invalid rows (neutralized > emissions or negative): {len(invalid)}")

# Check correlation
print("Correlation between emission and cost:", df.iloc[:, 0].corr(df.iloc[:, 1]))
print("Correlation between cost and neutralized:", df.iloc[:, 1].corr(df.iloc[:, 2]))
```
